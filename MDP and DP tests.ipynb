```python
from rl.model_based.environments.grid_world import GridWorld

import rl.model_based.prediction as pred
import rl.model_based.control as control

import rl.policy as policy
```

```python
world = GridWorld(5, 5, (1,2))
world.visualize()
```

```python
V = pred.evaluate_policy(policy.Deterministic.constant_policy(0), world)
world.visualize(V=V)
```

```python
V = pred.evaluate_policy(policy.Stochastic.uniform_policy(len(world.actions)), world, delta=0.01)
world.visualize(V=V)
```

```python
V = pred.value_iteration(world, printit=True)
```

```python
policy = control.optimal_policy(V, world)
world.visualize(policy=policy)
```

```python
policy = control.policy_iteration(world, start_uniform=True)
```

```python
policy = control.policy_iteration(world, start_uniform=False)
```

```python
policy = control.value_iteration(world)
```
